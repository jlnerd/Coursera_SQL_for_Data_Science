```python
import sqlite3
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

#creat engine
engine = create_engine('sqlite:///./dbs/chinook.db')

#save the table names to a list
table_names = engine.table_names()
#display(pd.DataFrame(table_names))

#Open engine conncetion
con = engine.connect()
#perform query
rs = con.execute("SELECT * FROM employees")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

```python
#Robert Kings Mailing Address:
pd.DataFrame(df.iloc[6])
```
