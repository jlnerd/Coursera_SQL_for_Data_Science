### Initialize SQL db

```python
import sqlite3
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

#creat engine
engine = create_engine('sqlite:///./dbs/chinook.db')

#save the table names to a list
table_names = engine.table_names()
display(pd.DataFrame(table_names, columns= ['Table Names']))
```

## 1
All of the questions in this quiz refer to the open source Chinook Database. Please familiarize yourself with the ER diagram to familiarize yourself with the table and column names to write accurate queries and get the appropriate answers.

Run Query: Find all the tracks that have a length of 5,000,000 milliseconds or more.

```python
#Open engine conncetion
con = engine.connect()
#get first 10 rows and headers
rs = con.execute("SELECT * FROM tracks LIMIT 5")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

```python
#Find all the tracks that have >5000000 mls length
rs = con.execute("SELECT * FROM tracks WHERE Milliseconds>5000000")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

How many tracks are returned?

```python
print("Number of tracks = ",df.shape[0])
```

## 7
Run Query: Find all the invoices from the billing city Brasilia, Edmonton, and Vancouver and sort in descending order by invoice ID.

```python
rs = con.execute("SELECT * \
                    FROM invoices \
                    where BillingCity IN ('Brasilia', 'Edmonton', 'Vancouver') \
                    ORDER BY InvoiceID DESC")
df = pd.DataFrame(rs.fetchall(),columns = rs.keys())
df
```

What is the total invoice amount of the first record returned? Enter the number below without a $ sign. Remember to sort in descending order to get the correct answer.

```python
df.iloc[0]
```

## 8
Run Query: Show the number of orders placed by each customer (hint: this is found in the invoices table) and sort the result by the number of orders in descending order.

```python
rs = con.execute("SELECT CustomerID, COUNT(InvoiceId) \
                    FROM invoices \
                    GROUP BY CustomerID \
                    ORDER BY COUNT(InvoiceId) DESC ")
df = pd.DataFrame(rs.fetchall(),columns = ['CustomerId','COUNT(CustomerId)'])
```

What is the number of items bought for the 8th person on this list? Enter that number below.

```python
df.iloc[8-1]
```

## 9
Run Query: Find the albums with 12 or more tracks.

```python
rs = con.execute("SELECT * \
                    FROM tracks \
                    GROUP BY AlbumID \
                    HAVING COUNT(DISTINCT TrackID)>=12")
df = pd.DataFrame(rs.fetchall(),columns = rs.keys())
df
```

While the number of records returned is limited to 10, the query, if run correctly, will indicate how many total records there are. Enter that number below.
