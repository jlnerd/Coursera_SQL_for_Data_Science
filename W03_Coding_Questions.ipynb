# Initialize

```python
import sqlite3
import pandas as pd
import numpy as np
from sqlalchemy import create_engine

#creat engine
engine = create_engine('sqlite:///./dbs/chinook.db')

#save the table names to a list
table_names = engine.table_names()
display(pd.DataFrame(table_names, columns= ['Table Names']))
```

# 1
Using a subquery, find the names of all the tracks for the album "Californication".

```python
#Open engine conncetion
con = engine.connect()
rs = con.execute("SELECT * \
                    FROM tracks TR \
                    CROSS JOIN albums ALB \
                    ON ALB.Albumid = TR.Albumid \
                    AND ALB.Title ='Californication'\
                ")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

What is the title of the 8th track?

```python
df.iloc[7]['Name']
```

# 2
Find the total number of invoices for each customer along with the customer's full name, city and email.

```python
con = engine.connect()
rs = con.execute("SELECT * \
                    FROM customers cust \
                    LEFT JOIN invoices inv \
                    ON cust.Customerid = inv.Customerid \
                ")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

# 3
Retrieve the track name, album, artistID, and trackID for all the albums.

```python
con = engine.connect()
rs = con.execute("SELECT tr.Name as 'Track_Name', alb.Title, art.Artistid, tr.Trackid \
                    FROM tracks tr \
                    LEFT JOIN albums alb \
                    ON tr.Albumid = alb.Albumid \
                    LEFT JOIN artists art \
                    ON art.Artistid = alb.Artistid \
                ")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

What is the song title of trackID 12 from the "For Those About to Rock We Salute You" album? Enter the answer below.

```python
df[df['Title']=='For Those About To Rock We Salute You']
```

# 5
Find the name and ID of the artists who do not have albums.

```python
con = engine.connect()
rs = con.execute("SELECT *\
                    FROM artists art \
                    LEFT JOIN albums alb \
                    ON art.Artistid = alb.Artistid \
                    WHERE alb.Albumid ISNULL \
")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

See if there are any customers who have a different city listed in their billing city versus their customer city.

```python
con = engine.connect()
rs = con.execute("SELECT *\
                    FROM customers cust \
                    LEFT JOIN invoices inv\
                    ON cust.Customerid = inv.Customerid \
                    WHERE  cust.City != inv.BillingCity \
")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```
