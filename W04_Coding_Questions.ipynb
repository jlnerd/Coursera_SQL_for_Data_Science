# 1
Pull a list of customer ids with the customer’s full name, and address, along with combining their city and country together. Be sure to make a space in between these two and make it UPPER CASE.

```python
#Open engine conncetion
con = engine.connect()
rs = con.execute("SELECT cust.CustomerID, cust.FirstName||' '||cust.LastName AS 'Full Name', cust.Address, cust.City, cust.Country, UPPER(cust.City||' '||cust.Country) AS 'City Country' \
                    FROM Customers cust \
                ")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

# 2
Create a new employee user id by combining the first 4 letter of the employee’s first name with the first 2 letters of the employee’s last name. Make the new field lower case and pull each individual step to show your work.

```python
con = engine.connect()
rs = con.execute("SELECT LOWER(SUBSTR(emp.FirstName,1,4)||SUBSTR(emp.LastName,1,2)) AS 'user id'\
                    FROM Employees emp \
                ")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

What is the final result for Robert King?

```python
df.iloc[6]
```

# 3
Show a list of employees who have worked for the company for 15 or more years using the current date function. Sort by lastname ascending.

```python
con = engine.connect()
rs = con.execute("SELECT *, DATE('now')-HireDate AS 'Years Employed' \
                    FROM Employees emp \
                ")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

What is the lastname of the last person on the list returned?


```python
df.iloc[df.shape[0]-1]['LastName']
```

# 5
Find the cities with the most customers and rank in descending order.

```python
con = engine.connect()
rs = con.execute("SELECT COUNT(CustomerID), City \
                    FROM Customers \
                    GROUP BY City \
                    ORDER BY COUNT(CustomerID) DESC \
")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```

# 6
Create a new customer invoice id by combining a customer’s invoice id with their first and last name while ordering your query in the following order: firstname, lastname, and invoiceID.

```python
con = engine.connect()
rs = con.execute("SELECT cust.FirstName||cust.LastName||inv.InvoiceID \
                    FROM customers cust \
                    LEFT JOIN Invoices inv \
                    ON cust.CustomerId = inv.CustomerId \
                    ORDER BY cust.FirstName, cust.LastName, inv.InvoiceID \
")
df =pd.DataFrame(rs.fetchall(), columns = rs.keys())
display(df)
```
